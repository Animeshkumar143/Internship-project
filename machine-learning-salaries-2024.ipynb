```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/salarie/salaries.csv")
```

```python
data.head(2)
```

```python
data.info()
```

```python
# Distribution of Salary by Experience Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=data)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Distribution of Salary by Employment Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=data)
plt.title('Salary Distribution by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Distribution of Remote Work Ratio
plt.figure(figsize=(10, 6))
sns.histplot(data['remote_ratio'], bins=20, kde=True)
plt.title('Distribution of Remote Work Ratio')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
# Salary Trends Over Years
plt.figure(figsize=(10, 6))
sns.lineplot(x='work_year', y='salary_in_usd', data=data)
plt.title('Salary Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Top 10 Job Titles with Highest Salaries
top_job_titles = data.groupby('job_title')['salary_in_usd'].median().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_job_titles.plot(kind='bar')
plt.title('Top 10 Job Titles with Highest Salaries')
plt.xlabel('Job Title')
plt.ylabel('Median Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Company Sizes
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='company_size', data=data, palette='viridis')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.title('Distribution of Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot for Numeric Variables
sns.pairplot(data[['salary_in_usd', 'remote_ratio', 'work_year']])
plt.show()
```

```python
# Distribution of Remote Work Ratio by Employment Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='remote_ratio', data=data)
plt.title('Distribution of Remote Work Ratio by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Remote Work Ratio')
plt.show()
```

```python
# Salary Distribution by Experience Level and Employment Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', hue='employment_type', data=data)
plt.title('Salary Distribution by Experience Level and Employment Type')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()
```

```python
# Salary Distribution by Company Size and Employment Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='company_size', y='salary_in_usd', hue='employment_type', data=data)
plt.title('Salary Distribution by Company Size and Employment Type')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()
```
